```python
import pandas as pd
import numpy as np
import os
```

```python
# 1. Verification Gate: Confirm the consolidated master table exists
master_file = "master_feature_table.csv"
if not os.path.exists(master_file):
    raise FileNotFoundError(
        f"Critical Ingestion Error: The consolidated sheet '{master_file}' "
        f"was not found. Please run Notebook 05A first."
    )
```

```python
# Load the unified dataset
master = pd.read_csv(master_file)
print("=====================================================================")
print(f"✅ SUCCESS: Consolidated Feature Table loaded from '{master_file}'")
print("=====================================================================")
print(f" -> Matrix Row Count:    {master.shape[0]}")
print(f" -> Matrix Column Count: {master.shape[1]}")
```

## Structural Integrity & Dimensional Assertions (Layers 1–5)

```python
print("--- INITIATING STRUCTURAL & CHRONOLOGICAL ASSERTION GATES ---\n")

# Validation Layer 1: Dataset dimensions check
assert len(master) == 18260, f"Dimension Error: Found {len(master)} rows instead of 18260."
print("  Layer 1 Passed: Dataset dimension matches exactly 18,260 daily observations.")
```

```python
# Validation Layer 2: Unique composite indexing primary keys
duplicate_keys_count = master.duplicated(["location", "date"]).sum()
assert duplicate_keys_count == 0, f"Key Collision Error: Found {duplicate_keys_count} duplicate rows."
print(" Layer 2 Passed: Composite primary keys [location, date] are completely unique.")
```

```python
# Validation Layer 3: Critical feature column null check
critical_cols = ["location", "date", "temp_max", "crowd_baseline", "budget_stress_index"]
for col in critical_cols:
    null_sum = master[col].isnull().sum()
    assert null_sum == 0, f"Data Leakage Error: Found {null_sum} missing fields in critical column '{col}'."
print(" Layer 3 Passed: Critical feature columns contain exactly zero null entries.")
```

```python
# Validation Layer 4: Date range temporal bounds verification
min_date_found = master["date"].min()
max_date_found = master["date"].max()
assert min_date_found == "2021-01-01", f"Timeline Error: Minimum date is {min_date_found} instead of 2021-01-01."
assert max_date_found == "2025-12-31", f"Timeline Error: Maximum date is {max_date_found} instead of 2025-12-31."
print(f" Layer 4 Passed: Temporal tracking bounds are strictly verified between {min_date_found} and {max_date_found}.")
```

```python
# Validation Layer 5: Destination cardinality verification
unique_destinations = master["location"].nunique()
assert unique_destinations == 10, f"Cardinality Error: Found {unique_destinations} locations instead of 10."
print(f" Layer 5 Passed: Destination geography count is verified at exactly {unique_destinations} unique cities.")
```

## Numeric Sanity Bounds & Physical Limits (Layer 6)

```python
print("--- INITIATING NUMERIC BOUNDARY VALUE ASSERTION GATES ---\n")

# Validation Layer 6: Lower and upper logical physics bounds checks
assert (master["crowd_baseline"] >= 0).all(), "Boundary Error: Found negative index in crowd_baseline."
assert (master["crowd_baseline"] <= 100).all(), "Boundary Error: Value in crowd_baseline exceeds maximum ceiling threshold of 100."

assert (master["budget_stress_index"] >= 0).all(), "Boundary Error: Found negative metric in budget_stress_index."
assert (master["budget_stress_index"] <= 100).all(), "Boundary Error: Value in budget_stress_index exceeds maximum ceiling threshold of 100."

assert (master["transport_complexity_score"] >= 0).all(), "Boundary Error: Found negative metric in transport_complexity_score."
assert (master["nearest_landslide_km"] >= 0).all(), "Boundary Error: Found negative distance scalar in nearest_landslide_km."

print(" ✅ Layer 6 Passed: Continuous numeric ranges and bounding values are perfectly sound.")
```

## Automated Dataset Health Report (Layer 7)

```python
# Validation Layer 7: Compile and render your stylized master feature health matrix
total_rows = len(master)
total_columns = master.shape[1]
total_locations = master["location"].nunique()
duplicate_keys = master.duplicated(["location", "date"]).sum()

# Critical null counting across our primary model columns
critical_cols = ["location", "date", "temp_max", "crowd_baseline", "budget_stress_index"]
critical_nulls = master[critical_cols].isnull().sum().sum()
```

```python
print("=====================================")
print("     MASTER DATASET HEALTH REPORT    ")
print("=====================================")
print(f"Rows:           {total_rows}")
print(f"Columns:        {total_columns}")
print(f"Locations:      {total_locations}")
print(f"Duplicate Keys: {duplicate_keys}")
print(f"Critical Nulls: {critical_nulls}")
print("\nSTATUS:")
print("✅ READY FOR RISK ENGINE")
print("=====================================")
```
